```python
from langchain_community.embeddings import DashScopeEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.example_selectors import SemanticSimilarityExampleSelector
import os
import dotenv

dotenv.load_dotenv()


# 2.å®šä¹‰åµŒå…¥æ¨¡å‹


os.environ["OPENAI_API_KEY"] = os.getenv("DASHSCOPE_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("DASHSCOPE_API_KEY")

# åˆ›å»ºé˜¿é‡Œäº‘çš„åµŒå…¥æ¨¡å‹å®ä¾‹
embeddings_model = DashScopeEmbeddings(
    model="text-embedding-v4",  # é€‰æ‹©é€‚å½“çš„æ¨¡å‹ï¼Œä¾‹å¦‚ text-embedding-v4
    dashscope_api_key=os.getenv("DASHSCOPE_API_KEY"),
)

# 3.å®šä¹‰ç¤ºä¾‹ç»„
examples = [
    {"input": "é«˜å…´", "output": "æ‚²ä¼¤"},
    {"input": "é«˜", "output": "çŸ®"},
    {"input": "é•¿", "output": "çŸ­"},
    {"input": "ç²¾åŠ›å……æ²›", "output": "æ— ç²¾æ‰“é‡‡"},
    {"input": "é˜³å…‰", "output": "é˜´æš—"},
    {"input": "ç²—ç³™", "output": "å…‰æ»‘"},
    {"input": "å¹²ç‡¥", "output": "æ½®æ¹¿"},
    {"input": "å¯Œè£•", "output": "è´«ç©·"},
]

example_selector = SemanticSimilarityExampleSelector.from_examples(
    examples,
    embeddings_model,
    FAISS,
    k=2,
)

example_prompt = PromptTemplate(
    # input_variables=["input", "output"],
    template="Input: {input}\nOutput: {output}",  # å®šä¹‰å¦‚ä½•æ ¼å¼åŒ–æ¯ä¸ªç¤ºä¾‹
)


similar_prompt = FewShotPromptTemplate(
    example_selector=example_selector,
    example_prompt=example_prompt,
    prefix="ç»™å‡ºæ¯ä¸ªè¯ç»„çš„åä¹‰è¯",
    suffix="Input: {word}\nOutput:",
    input_variables=["word"],
)

response = similar_prompt.invoke({"word":"å¿§éƒ"})
print(response.text)
```
